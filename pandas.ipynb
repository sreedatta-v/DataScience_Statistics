```python
"""
Data Cleaning
-> Deleting certain Rows or Coloums which contains wrong informations or outliers or Null values.
-> Bad Data Could be:
• Empty cells
• Data in wrong format
• Wrong data
• Duplicates

Empty Cells can potentially give wrong cells when we analyse tehe data.
"""
#Removing Rows

import pandas as pd
df = pd.read_csv('data.csv')
print(df.head(20))
new_df = df.dropna() #dropna() basically returns the new DataFrame and it doesnt affect old Frame
# To change old DataFrame use inplace = True "df.dropna(inplace = True)"
print(new_df.head(20))
```

```python
#Replacing Empty Cells
# This method helps in filling rows insteading of deleting them.

import pandas as pd
df = pd.read_csv('data.csv')
df.fillna(21, inplace = True) #here the place 21, denotes the value to be filled.
print(df.head(20))
#For Filling only certain column we basically use,
df['Calories'].fillna(130, inplace = True)
```

```python
#Replacing using  Mean, Median, Mode
# It is a common method to replace the empty cells.

import pandas as pd
df = pd.read_csv('data.csv')
mean_data = df['Calories'].mean()
df['Calories'].fillna(mean_data, inplace = True)
print(df.head(20))
#Similarly used for median, mode
```

```python
#Cleaning data of wrong format
#basically to fix it, we have two options either to delete the row or to convert cells in the column to correct format.

import pandas as pd
df = pd.read_csv('data.csv')
df['Duration'] = pd.to_numeric(df['Duration'], errors = 'coerce') #coerce is used to replace the wrong format.
```
