```python
import numpy as np
import pandas as pd
import scipy.stats as stats
from scipy.stats import ttest_1samp
```

```python
data = pd.read_csv("onesample.csv")
```

```python
data.head()
```

```python
data = data['Insect Length']
```

```python
data.head()
```

```python
mean = data.mean()
print(mean)
```

```python
t_statistic,p_value = ttest_1samp(data, 6.09)
#6.09 is the hypothesized Mean
# test statistic is a measure of how far the sample mean is from hypothesized mean.
# p value, assuming null hypothesis is True, probability of observing a test statistics as extreme as one calc from sample.
```

```python
print(t_statistic)
print(p_value)
```
