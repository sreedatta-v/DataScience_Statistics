```python
cor = data.corr() #Calcs the value of correlation matrix
# Corelation matrix is a square matrix that contains the correlation coefficients between each pair of variables in DF
plt.figure(figsize=(16,10))
sns.heatmap(cor)
"""
-1: Perfect negative correlation
0: No correlation
1: Perfect positive correlation

-> Heatmap is a graphical representation of matrix where values are in colors.
-> Colored according to the Strength
-> Seaborn library is used for visualization
-> Red indicates a strong positive, blue indicates strong negative.
"""
```
