```python
import pandas as pd
import numpy as np
import scipy.stats as stats
from scipy.stats import chisquare
```

```python
data = pd.read_csv("chi_square.csv")
#The purpose of the chi-square test in statistics is to determine whether there is a significant difference between the expected frequencies and the observed frequencies in one or more categories.
# It is used to find the independence of two categorical variables.
```

```python
data
```

```python
t_statistic, p_value = chisquare(f_obs= data['Observed'],f_exp=data['Expected'])
"""
-> f_obs: it is a list of observed frequency
-> f_exp: It is a list of expected frequency
-> t_statistics, p_value are both used for outputs.
-> t_statistics is a difference between observed and expected frequency
-> p_value is a probability of obtaining a chisquare statistics as extreme as, assuming that the null hypothesis is true.
"""
```

```python
p_value
#As p > 0.05, it falls under null hypotesis, which means observed value = predicted value
```
